```python
import pandas as pd
import shap
from sklearn.impute import KNNImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from econml.grf import CausalForest
```

```python
data = pd.read_csv("bsc_project_set.csv", index_col=0)

xs = data.copy()

y = xs["mort_28"]
w = xs["peep_regime"]

w = pd.Series(map(lambda x: 0 if x == "low" else 1, w))
y = pd.Series(map(lambda x: 1 if x == False else 0, y))

xs = xs.drop(columns=["id", "mort_28", "peep", "peep_regime"])
columns_x = xs.columns

xs.loc[xs["sex"] == "M", "sex"] = 0
xs.loc[xs["sex"] == "F", "sex"] = 1

norm_xs = (xs - xs.mean())/xs.std()

imputer = KNNImputer(n_neighbors=2, weights="uniform")
imp_xs = imputer.fit_transform(norm_xs)

imp_xs = pd.DataFrame(data=imp_xs, columns=columns_x)

imp_xs
```

```python
full_data = imp_xs.assign(W=w, Y=y)
```

```python
train, test = train_test_split(full_data, test_size=0.3)
train_x = train[columns_x]
test_x = test[columns_x]
```

```python
# S-learner
regr_s = RandomForestRegressor()
regr_s.fit(train.drop(columns=["Y"]), train["Y"])

# T-learner
regr0_t = RandomForestRegressor()
regr1_t = RandomForestRegressor()
regr0_t.fit(train.query("W==0").drop(columns=["W", "Y"]), train.query("W==0")["Y"])
regr1_t.fit(train.query("W==1").drop(columns=["W", "Y"]), train.query("W==1")["Y"])

# Causal forest
causal_forest = CausalForest()
causal_forest.fit(X=train_x, T=train["W"], y=train["Y"])
```

# S-learner

```python
explainer_s = shap.Explainer(model=(lambda xs: regr_s.predict(xs.assign(**{"W": 1})) - regr_s.predict(xs.assign(**{"W": 0}))), masker=shap.utils.sample(train_x, 100))
shap_values_s = explainer_s(imp_xs)
shap.plots.bar(shap_values_s, max_display=24)
```

```python
explainer_s = shap.Explainer(model=(lambda xs: regr_s.predict(xs.assign(**{"W": 1})) - regr_s.predict(xs.assign(**{"W": 0}))), masker=shap.maskers.Independent(train_x, max_samples=100))
shap_values_s = explainer_s(imp_xs)
shap.plots.bar(shap_values_s, max_display=24)
```

```python
shap.plots.beeswarm(shap_values_s, max_display=23)
```

# T-leaner

```python
explainer_t = shap.Explainer(model=(lambda xs: regr1_t.predict(xs) - regr0_t.predict(xs)), masker=shap.utils.sample(train_x, 100))
shap_values_t = explainer_t(imp_xs)
shap.plots.bar(shap_values_t, max_display=24)
```

```python
explainer_t = shap.Explainer(model=(lambda xs: regr1_t.predict(xs) - regr0_t.predict(xs)), masker=shap.maskers.Independent(train_x, max_samples=100))
shap_values_t = explainer_t(imp_xs)
shap.plots.bar(shap_values_t, max_display=24)
```

```python
shap.plots.beeswarm(shap_values_t, max_display=23)
```

# Causal Forest

```python
explainer_forest = shap.Explainer(causal_forest.predict, shap.utils.sample(train_x, 100))
shap_values_forest = explainer_forest(imp_xs)
shap.plots.bar(shap_values_forest, max_display=23)
```

```python
shap.plots.beeswarm(shap_values_forest, max_display=23)
```
